#### Problema 1.

Escribe un algoritmo que permita calcular el factorial $$n!=n\cdot(n-1)\cdot(n-2)\cdot\cdots\cdot(2)(1)$$ de un número $n$ dado por el usuario.

*Solución*.

*Paso 1*. Pedir un número entero $n$ positivo.

```python
# Escribe tu código aquí.

def factorial(n):

    if n == 0 or n == 1:
        print(f"{n}! = 1")
        return 1
    else:
        recursivo = factorial(n-1)
        resultado = n * recursivo
        print(f"{n}! = {resultado}")
        return resultado

def main():
    try:
        n = int(input("Introduce un número entero positivo: "))
        if n < 0:
            print("El factorial no está definido para números negativos.")
            return
    except ValueError:
        print("¡Oye! Eso no es un número entero.")
        return

    resultado = factorial(n)
    print(f"El factorial de {n} es: {resultado}")

main()
```

#### Problema 2.
Escribe un programa que permita determinar si un número $n$ dado por el usuario es primo o no lo es.

*Solución*.
*Paso 1*. Pedir un número entero $n$ positivo.

```python
# Escribe tu código aquí.

def es_primo(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limite = int(n**0.5) + 1
    for i in range(3, limite, 2):
        if n % i == 0:
            return False
    return True

try:
    numero = int(input("Ingresa un número: "))
except ValueError:
    print("¡Oye! Eso no es un número entero.")

if es_primo(numero):
    print(f"El número {numero} es primo.")
else:
    print(f"El número {numero} no es primo.")
```

#### Problema 3.
Escribe un algoritmo que permita calcular el máximo común divisor de dos números enteros positivos dados por el usuario.

*Solución*.

```python
# Escribe tu código aquí.

def mcd_while(a, b):
    while b != 0:
        a, b = b, a % b
        print(a,b)
    return a

def main():

    try:
        a = int(input("Introduce el primer número: "))
        b = int(input("Introduce el segundo número: "))
    except ValueError:
        print("¡Oye! Eso no es un número entero.")
        return

    resultado = mcd_while(a, b)
    print(f"El MCD de {a} y {b} es: {resultado}")

main()
```

#### Problema 4.
Escribe un algoritmo que permita averiguar si dos números enteros positivos dados por el usuario son primos relativos o no lo son.

*Solución*.

```python
# Escribe tu código aquí.

def mcd_while(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def main():

    try:
        a = int(input("Introduce el primer número: "))
        b = int(input("Introduce el segundo número: "))
    except ValueError:
        print("¡Oye! Eso no es un número entero.")
        return

    resultado = mcd_while(a, b)
    print(f"El MCD de {a} y {b} es: {resultado}")

    if resultado == 1:
        print("Los números son primos relativos.")
    else:
        print("Los números NO son primos relativos.")

main()
```

#### Problema 5.
Escribe un algoritmo que permita encontrar los factores de un número entero dado por el usuario.

*Solución*.

```python
# Escribe tu código aquí.

def factores(n):
    lista_factores = []
    for i in range(1, abs(n) + 1):
        if n % i == 0:
            lista_factores.append(i)
            lista_factores.append(-i)
    return lista_factores

def main():
    try:
        numero = int(input("Introduce un número entero: "))
    except ValueError:
        print("¡Oye! Eso no es un número entero.")
        return

    resultado = factores(numero)
    print(f"Los factores de {numero} son: {resultado}")

main()
```
